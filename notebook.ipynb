## Paso 2 — Configurar el `.env`

Crea un archivo `.env` en la raíz del proyecto y completa estos valores:

```
OPENAI_API_KEY=sk-...
MONGODB_CONNECTION_STRING=mongodb+srv://...
TELEGRAM_BOT_TOKEN=        <-- lo obtienes en el Paso 4
SUPABASE_URL=https://xxxx.supabase.co
SUPABASE_KEY=sb_publishable_...
```

> **Nunca subas el `.env` a GitHub.** Verifica que está en el `.gitignore`.

````python
from dotenv import load_dotenv
import os

load_dotenv()

variables = ["OPENAI_API_KEY", "MONGODB_CONNECTION_STRING", "TELEGRAM_BOT_TOKEN", "SUPABASE_URL", "SUPABASE_KEY"]

for var in variables:
    estado = "OK" if os.getenv(var) else "FALTA"
    print(f"{estado} — {var}")
````

## Paso 3 — Conectar a Supabase y crear las tablas

1. Crea un proyecto en [supabase.com](https://supabase.com)
2. Ve a **Settings → API** y copia la **URL** y la **service_role key**
3. Ve a **SQL Editor → New query**, pega el contenido de `schema.sql` y ejecútalo

````python
from supabase import create_client
import os
from dotenv import load_dotenv

load_dotenv()
client = create_client(os.getenv("SUPABASE_URL"), os.getenv("SUPABASE_KEY"))

tablas = ["sessions", "messages", "traces", "model_pricing", "costs"]

for tabla in tablas:
    try:
        client.table(tabla).select("*").limit(1).execute()
        print(f"OK — {tabla}")
    except Exception as e:
        print(f"ERROR — {tabla}: {e}")
        print("  -> Ejecuta el schema.sql en el SQL Editor de Supabase")
````

## Paso 4 — Crear tu bot en BotFather y obtener el token

1. Abre Telegram y busca **@BotFather**
2. Escribe `/newbot`
3. Elige un nombre (ej: `Asistente DuocUC`)
4. Elige un username que termine en `bot` (ej: `asistente_duocuc_bot`)
5. BotFather te entrega un token — cópialo en el `.env` como `TELEGRAM_BOT_TOKEN=...`

````python
import httpx, os
from dotenv import load_dotenv

load_dotenv()

token = os.getenv("TELEGRAM_BOT_TOKEN")
response = httpx.get(f"https://api.telegram.org/bot{token}/getMe").json()

if response["ok"]:
    bot = response["result"]
    print(f"Bot activo: {bot['first_name']} (@{bot['username']})")
else:
    print("Token invalido — revisa el .env")
````
